**[*args --Variable positional arguments]**

```python
def sum(*args):
    total=0
    for i in args:
        total+=i
    return total
```

```python
sum(1,2,3,4,5)
```

**[**kwargs variable keyword arguments]**

```python
def example(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}:{value}")
```

```python
example(Name="Sricharan",Age=22,Role="Developer")
```

**Unpacking arguments**

```python
def add_numbers(a,b,c):
    return a+b+c
numbers = [1,2,3]
```

```python
add_numbers(*numbers)
```

**lambda function and Map(),Reduce(),filter()**

**map()**

```python
list1 = [1,3,4,5]
sqr_lst = []
def square(list1):        #without map function
    for i in list1:
        sqr_lst.append(i**2)
    return sqr_lst
        
        
```

```python
sqr_lst
```

```python
nums = [2,3,5,6,7]
sqr_list = list(map(lambda x:x**2,nums)) #with map function
```

```python
sqr_list
```

**Filter()**

```python
scores = {'Math':90,'Science':85,'English':88}
result = dict(filter(lambda x:x[0]=="English",scores.items()))  #filter function to filter subjects with scores >=88
```

```python
result
```

```python
res = []
def do(scores):
    for key, value in scores.items():
        if value>=88:
            res.append((key,value))
```

```python
do(scores)
```

```python
res
```

```python
dict(res)
```
